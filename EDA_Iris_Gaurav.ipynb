```python
# Data manipulation and analysis
import pandas as pd

# Data visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Set default plot style
sns.set(style="whitegrid")
```

```python
# Load the built-in Iris dataset from Seaborn
df = sns.load_dataset('iris')

# Show the first few rows
df.head()
```

```python
# Get the structure of the dataset
df.info()
```

```python
# Statistical summary of the dataset
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Distribution of Sepal Length
sns.histplot(df['sepal_length'], kde=True)
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
# Distribution of Petal Width
sns.histplot(df['petal_width'], kde=True, color='orange')
plt.title("Distribution of Petal Width")
plt.show()
```

```python
# Scatterplot for Petal Length vs Petal Width
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.title("Petal Length vs Width by Species")
plt.show()
```

```python
# Boxplot to compare Sepal Length across species
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title("Sepal Length Across Species")
plt.show()
```

```python
# Pairplot to visualize pairwise relationships
sns.pairplot(df, hue='species')
plt.suptitle("Pairwise Plot of All Features", y=1.02)
plt.show()
```

```python
# Drop non-numeric column before calculating correlation
numeric_df = df.drop(columns=['species'])

# Correlation heatmap
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()
```

### Key Insights

- **Petal length** and **petal width** show strong correlation and are effective in separating species.
- **Setosa** species is clearly distinct from the other two based on petal size.
- **Versicolor** and **Virginica** show some overlap but can still be distinguished with multiple features.
- No missing data; the dataset is clean and ready for modeling.
